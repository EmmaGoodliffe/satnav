To create necessary environment:

1. Install `conda`
1. Create and activate a new environment: `conda create -n pyrosmnx` and `conda activate pyrosmnx`
1. Set the python version to 3.8: `conda install python=3.8`
1. Install old versions of `shapely`: `conda install main::shapely=1.8.4`
1. Install `pyrosm` and `osmnx`: `conda install conda-forge::pyrosm conda-forge::osmnx`

```python
import pyrosm
import osmnx
```

```python
osm = pyrosm.OSM("monaco.osm.pbf")
```

```python
# get full road network (can take a while!)
nodes, edges = osm.get_network(network_type="driving", nodes=True)
```

```python
# convert to OSMnx graph
G = osmnx.graph_from_gdfs(nodes, edges)
```

```python
# save for later
osmnx.save_graphml(G, "england.graphml")
```
