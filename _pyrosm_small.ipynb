# Show map downloaded from [geofabrik](https://download.geofabrik.de/europe/great-britain.html)

```python
# requires conda env ∵ of C++ stuff I think
import numpy as np
import pyrosm
from matplotlib import pyplot as plt
```

```python
# coord = np.array([43.738899, 7.426980][::-1])
# osm = pyrosm.OSM("monaco.osm.pbf",bounding_box=list(coord) + list(coord+0.01) )

coord = np.array([51.754179, -1.258663][::-1])
osm = pyrosm.OSM("england.osm.pbf",bounding_box=list(coord) + list(coord+0.01) )
```

```python
roads = osm.get_network(network_type="driving")
```

```python
fig, ax = plt.subplots()
roads.plot(ax=ax, linewidth=0.8, color="black")
# ax.set_axis_off()
# plt.savefig("map.png", dpi=200, bbox_inches="tight")
plt.show()
plt.close()
```
